```python
# GRÁFICO DE DISPERSÃO COM LEGENDAS p2

import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

```python
# Carregamento dos dados

base = pd.read_csv('co2.csv')
base.head()


# Obs: A idéia é criar um gráfico de dispersão usando em que os eixos x e y são referentes aos dados das colunas, conc e uptake
# De modo que sejam plotados os pontos em que são caracterizados pelo tratamento nonchilled e chilled, com cores diferentes, 
# e label/legenda apontando para os tipos de tratamento, no canto inferior direito da tela
```

```python
base.shape
```

```python
# Criar duas variáveis para cada atributo (x = conc e y = uptake)


x = base.conc
y = base.uptake
x
```

```python
# Retornar os valores únicos do atributo 'Treatment'
# OBS: Reparar na fórmula utilizada para criar uma lista, com apenas valores únicos, ou seja, não reptidos.


unicos = list(set(base.Treatment))
unicos
```

```python
# Percorrer cada tipo de tratamento (chilled e nonchilled)
# E criar um gráfico de dispersão


for i in range(len(unicos)):
    indice = (base.Treatment == unicos[i])
    plt.scatter(np.array(x)[indice], np.array(y)[indice], label=unicos[i])

plt.legend(loc = 'lower right')
plt.show()
```

```python
# Importação das bibliotecas

import seaborn as srn
import pandas as pd
import matplotlib.pyplot as plt
```

```python
# Carregamento da base de dados (co2)

base = pd.read_csv('co2.csv')
base.head()
```

```python
# Gráficos de dispersão utilizando os atributos conc e uptake
# Agrupamento pelo type


srn.scatterplot(data=base, x='conc', y='uptake', hue='Type')

# Obs: "  hue "
# O parâmetro hue da função scatterplot() do módulo seaborn é usado para especificar uma variável categórica que..
# ..será usada para colorir os pontos do gráfico de dispersão. O valor padrão é None, o que significa que todos..
# ..os pontos serão plotados na mesma cor.
```

```python
# Seleção de registros específicos da base de dados(Quebec e Mississipi)


q = base.loc[base['Type'] == 'Quebec']
m = base.loc[base['Type'] == 'Mississippi']
```

```python
# Subgráfico (1 x 2) mostrando gráficos sobre cada região
# dados = q e m
# atributos = conc e uptake
# Setar título


plt.figure()
plt.subplot(1, 2, 1)
srn.scatterplot(data=q, x='conc', y='uptake').set_title('Quebec')
plt.subplot(1, 2, 2)
srn.scatterplot(data=m, x='conc', y='uptake').set_title('Mississippi')
plt.tight_layout()
```

```python
base.head()
```

```python
# Criar objetos para os dados em Treatment - Refrigerado e não refrigerado


ch = base.loc[base['Treatment'] == 'chilled']
nc = base.loc[base['Treatment'] == 'nonchilled']
```

```python
ch.head()
```

```python
# # Subgráfico (1 x 2) usando os dados em Treatment
# dados = ch e nc
# atributos = conc e uptake
# Setar título


plt.figure()
plt.subplot(1, 2, 1)
srn.scatterplot(data=ch, x='conc', y='uptake').set_title('chilled')
plt.subplot(1, 2, 2)
srn.scatterplot(data=nc, x='conc', y='uptake').set_title('nonchilled')
plt.tight_layout()
```

```python
# GRÁFICOS 3D p6
# Importação das bibliotecas

import pandas as pd
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import axes3d
```

```python
# Carregamento da base de dados

base = pd.read_csv('orchard.csv')
base.shape
```

```python
# Criação de gráfico 3D, 
# Indicando o atributo projection = '3d'
# E passando 3 atributos (decrease, rowpos, colpos)


figura = plt.figure()
eixo = figura.add_subplot(1, 1, 1, projection='3d')


eixo.scatter(base.decrease, base.rowpos, base.colpos)
eixo.set_title('Gráfico 3D')
eixo.set_xlabel('decrease')
eixo.set_ylabel('rowpos')
eixo.set_zlabel('colpos')


plt.show()
```
